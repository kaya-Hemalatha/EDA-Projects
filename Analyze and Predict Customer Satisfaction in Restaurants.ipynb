```python
#1 customer satisfaction vary with different age groups
# Define age groups
age_bins = [18, 24, 34, 44, 54, 64, 100]
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
my_data['AgeGroup'] = pd.cut(my_data['Age'], bins=age_bins, labels=age_labels)

# Calculate average satisfaction ratings for each age group
satisfaction_by_age = my_data.groupby('AgeGroup', observed=False)['ServiceRating'].mean()

# Plot the results
satisfaction_by_age.plot(kind='bar',color = 'orange')
plt.title('Average Service Rating by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Service Rating')
plt.show()
```

```python
#2 cuisine is preferred by customers with high satisfaction
satisfaction = my_data[my_data['HighSatisfaction'] == 1]

# Calculate the distribution of preferred cuisines
cuisine_distribution = satisfaction['PreferredCuisine'].value_counts()

# Plot the results
cuisine_distribution.plot(kind='barh', color = 'pink')
plt.title('Preferred Cuisine of Highly Satisfied Customers')
plt.xlabel('Preferred Cuisine')
plt.ylabel('Number of Customers')
plt.show()
```

```python
#3 Frequency visits impact customer satisfaction
# Check the distribution of VisitFrequency
print(my_data['VisitFrequency'].value_counts())

# Group by VisitFrequency and calculate the mean satisfaction
satisfaction_by_frequency = my_data.groupby('VisitFrequency')['HighSatisfaction'].mean().reset_index()

# Bar plot of average satisfaction by visit frequency
sns.barplot(x='VisitFrequency', y='HighSatisfaction', data=satisfaction_by_frequency)
plt.title('Average Satisfaction by Visit Frequency')
plt.xlabel('Visit Frequency')
plt.ylabel('Average Satisfaction')
plt.show()

# Bar plot of distribution satisfaction within each visit frequency
sns.barplot(x='VisitFrequency', y='HighSatisfaction', data=my_data)
plt.title('Distribution of Satisfaction by Visit Frequency')
plt.xlabel('Visit Frequency')
plt.ylabel('Satisfaction (High/Low)')
plt.show()
```

```python
#4 time of visit affect customer satisfaction
# Check the distribution of TimeOfVisit
print(my_data['TimeOfVisit'].value_counts())

# Group by TimeOfVisit and calculate the mean satisfaction
satisfaction_by_time = my_data.groupby('TimeOfVisit')['HighSatisfaction'].mean().reset_index()

# Bar plot of average satisfaction by time of visit
sns.barplot(x='TimeOfVisit', y='HighSatisfaction', data=satisfaction_by_time)
plt.title('Average Satisfaction by Time of Visit')
plt.xlabel('Time of Visit')
plt.ylabel('Average Satisfaction')
plt.show()

# Bar plot to see the distribution of satisfaction within each time of visit
sns.barplot(x='TimeOfVisit', y='HighSatisfaction', data=my_data)
plt.title('Distribution of Satisfaction by Time of Visit')
plt.xlabel('Time of Visit')
plt.ylabel('Satisfaction (High/Low)')
plt.show()
```

```python
#5 customer dining for causal, business, or celebration are more satisfied
#check the disturbution of DiningOccasion
print(my_data['DiningOccasion'].value_counts())

#group by DiningOccasion and calculate the mean od satisfaction
satisfaction_by_occasion = my_data.groupby('DiningOccasion')['HighSatisfaction'].mean().reset_index()

# Bar plot of average satisfaction by dining occasion
sns.barplot(x='DiningOccasion', y='HighSatisfaction', data=satisfaction_by_occasion)
plt.title('Average Satisfaction by Dining Occasion')
plt.xlabel('Dining Occasion')
plt.ylabel('Average Satisfaction')
plt.show()

# Bar plot to see the distribution of satisfaction within each dining occasion
sns.barplot(x='DiningOccasion', y='HighSatisfaction', data=my_data)
plt.title('Distribution of Satisfaction by Dining Occasion')
plt.xlabel('Dining Occasion')
plt.ylabel('Satisfaction (High/Low)')
plt.show()
```

```python
#6 online reservations report higher satisfaction
# Group by 'OnlineReservation' and calculate the mean of satisfaction ratings
satisfaction_by_reservation = my_data.groupby('OnlineReservation')[['ServiceRating', 'FoodRating', 'AmbianceRating']].mean()
print(satisfaction_by_reservation)

# Plot Service Rating by Online Reservation
plt.figure(figsize=(5, 5))
sns.boxplot(x='OnlineReservation', y='ServiceRating', data=my_data)
plt.title('Service Rating by Online Reservation')
plt.xlabel('Online Reservation')
plt.ylabel('Service Rating')
plt.show()

# Plot Food Rating by Online Reservation
plt.figure(figsize=(5, 5))
sns.boxplot(x='OnlineReservation', y='FoodRating', data=my_data)
plt.title('Food Rating by Online Reservation')
plt.xlabel('Online Reservation')
plt.ylabel('Food Rating')
plt.show()

# Plot Ambiance Rating by Online Reservation
plt.figure(figsize=(5, 5))
sns.boxplot(x='OnlineReservation', y='AmbianceRating', data=my_data)
plt.title('Ambiance Rating by Online Reservation')
plt.xlabel('Online Reservation')
plt.ylabel('Ambiance Rating')
plt.show()
```

```python
#7 loyalty program members more satisified than non-members
# Plot Service Rating
plt.figure(figsize=(5, 5))
sns.boxplot(x='LoyaltyProgramMember', y='ServiceRating', data=my_data)
plt.title('Service Rating by Loyalty Program Membership')
plt.xlabel('Loyalty Program Member')
plt.ylabel('Service Rating')
plt.show()

# Plot Food Rating
plt.figure(figsize=(5, 5))
sns.boxplot(x='LoyaltyProgramMember', y='FoodRating', data=my_data)
plt.title('Food Rating by Loyalty Program Membership')
plt.xlabel('Loyalty Program Member')
plt.ylabel('Food Rating')
plt.show()

# Plot Ambiance Rating
plt.figure(figsize=(5, 5))
sns.boxplot(x='LoyaltyProgramMember', y='AmbianceRating', data=my_data)
plt.title('Ambiance Rating by Loyalty Program Membership')
plt.xlabel('Loyalty Program Member')
plt.ylabel('Ambiance Rating')
plt.show()
```
