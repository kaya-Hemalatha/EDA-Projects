```python
#1 rentals vary across different days of the month
# Convert 'datetime' column to datetime object
day['dteday'] = pd.to_datetime(day['dteday'])

# Extract day of the month
day['day_of_month'] = day['dteday'].dt.day

# Group by day of the month and sum up the total rentals
rentals_summary = day.groupby('day_of_month')['cnt'].sum().reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
plt.scatter(rentals_summary['day_of_month'], rentals_summary['cnt'], color='lightpink')
plt.xlabel('Day of the Month')
plt.ylabel('Total Rentals')
plt.title('Bike Rentals Across Different Days of the Month')
plt.xticks(range(1, 32))  # Ensure all days of the month are shown
plt.show()
```

```python
#2 weather condition that impacts the bike rentals
extreme_weather = day[day['weathersit'].isin([3, 4])]
extreme_rentals = extreme_weather.groupby('weathersit')['cnt'].sum().reset_index()

normal_weather = day[day['weathersit'].isin([1, 2])]
normal_rentals = normal_weather.groupby('weathersit')['cnt'].sum().reset_index()

# Combine both dataframes for a complete comparison
combined_dataframes = pd.concat([extreme_rentals, normal_rentals]) #concatenates and combines multiple DataFrames or Series into a single, unified DataFrame or Series

# Plot the results
plt.figure(figsize=(8, 5))
plt.plot(combined_dataframes['weathersit'], combined_dataframes['cnt'], marker='o', linestyle='--', color='purple')
plt.xlabel('Weather Situation')
plt.ylabel('Total Rentals')
plt.title('Impact of Extreme Weather Conditions on Bike Rentals - Line Plot')
plt.xticks(ticks=[1, 2, 3, 4], labels=['Clear/Cloudy', 'Mist', 'Light Snow/Rain', 'Heavy Rain/Snow'])
plt.grid(True)
plt.show()
```

```python
#3 Compare bike rentals on working days versus non-working days 
workingday_summary = day.groupby('workingday')['cnt'].sum().reset_index()

weekday_summary = day.groupby('weekday')['cnt'].sum().reset_index()

combined_summary = day.groupby(['workingday', 'weekday'])['cnt'].sum().reset_index()

pivot_summary = combined_summary.pivot(index='weekday', columns='workingday', values='cnt')

# Plotting the combined results
pivot_summary.plot(kind='bar', figsize=(10, 6), color=['lightcoral', 'lightblue'])
plt.xlabel('Day of the Week')
plt.ylabel('Total Rentals')
plt.title('Bike Rentals on Working Days vs Non-Working Days Across the Week')
plt.xticks(ticks=range(7), labels=['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
plt.legend(['Non-Working Day', 'Working Day'], title='Day Type') #clear visualization by telling the functionality of plot elements
plt.show()
```

```python
#4 holidays impact bike rentals
# Calculate the total rentals for holidays and non-holidays
holiday_totals = day.groupby('holiday')['cnt'].sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(holiday_totals, labels=['Non-Holiday', 'Holiday'], colors=['lightblue', 'green'], autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Bike Rentals on Holidays vs Non-Holidays')
plt.show()
```

```python
#5 demand for bikes during peak hours, weekdays, or weekends?
# Analyze peak hours
peak_hour_demand = hour.groupby('hr')['cnt'].mean()
print("Average bike demand per hour:", peak_hour_demand)

# Analyze weekdays vs weekends
hour['is_weekend'] = hour['weekday'].apply(lambda x: 1 if x == 0 or x == 6 else 0)
weekday_vs_weekend_demand = hour.groupby('is_weekend')['cnt'].mean()
print("\nAverage bike demand on weekdays vs weekends:", weekday_vs_weekend_demand)

# Combine peak hours with weekdays/weekends
peak_weekday_demand = hour[hour['is_weekend'] == 0].groupby('hr')['cnt'].mean()
peak_weekend_demand = hour[hour['is_weekend'] == 1].groupby('hr')['cnt'].mean()

print("\nAverage bike demand per hour on weekdays:",peak_weekday_demand)

print("\nAverage bike demand per hour on weekends:",peak_weekend_demand)

# Line plot for peak hours
plt.figure(figsize=(10, 6))
plt.plot(peak_hour_demand, marker='o')
plt.title('Average Bike Demand by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Bike Demand')
plt.grid(True)
plt.show()

# Line plot for weekdays vs weekends
plt.figure(figsize=(10, 6))
plt.plot(peak_weekday_demand, marker='o', label='Weekdays')
plt.plot(peak_weekend_demand, marker='o', label='Weekends')
plt.title('Average Bike Demand by Hour (Weekdays vs Weekends)')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Bike Demand')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#6 bike rentals vary across different seasons
# Map season codes to actual season names in both datasets
season_mapping = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'}
day['season'] = day['season'].map(season_mapping) #you can quickly convert the codes into human-readable labels, which enhances the clarity of your data analysis and visualizations.
hour['season'] = hour['season'].map(season_mapping)

# Analyze bike rentals across different seasons in the 'day' dataset
# Group by season and calculate the average bike rentals
seasonal_demand_day = day.groupby('season')['cnt'].mean()

# Analyze bike rentals across different seasons in the 'hour' dataset
# Group by season and calculate the average bike rentals
seasonal_demand_hour = hour.groupby('season')['cnt'].mean()

# Create a bar plot for average bike rentals by season from the 'day' dataset
plt.figure(figsize=(8, 5))
seasonal_demand_day.plot(kind='bar', color=['skyblue', 'lightgreen', 'orange', 'lightcoral'])
plt.title('Average Bike Rentals by Season (Day Dataset)')
plt.xlabel('Season')
plt.ylabel('Average Bike Rentals')
plt.xticks(rotation=0)
plt.show()

# Create a bar plot for average bike rentals by season from the 'hour' dataset
plt.figure(figsize=(8, 5))
seasonal_demand_hour.plot(kind='bar', color=['skyblue', 'lightgreen', 'orange', 'lightcoral'])
plt.title('Average Bike Rentals by Season (Hour Dataset)')
plt.xlabel('Season')
plt.ylabel('Average Bike Rentals')
plt.xticks(rotation=0)
plt.show()

# Create a box plot for distribution of bike rentals by season from the 'day' dataset
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='cnt', data=day, palette='Set2')
plt.title('Distribution of Bike Rentals by Season (Day Dataset)')
plt.xlabel('Season')
plt.ylabel('Bike Rentals')
plt.show()

# Create a box plot for distribution of bike rentals by season from the 'hour' dataset
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='cnt', data=hour, palette='Set2')
plt.title('Distribution of Bike Rentals by Season (Hour Dataset)')
plt.xlabel('Season')
plt.ylabel('Bike Rentals')
plt.show()
```

```python
#7 days of the week when bike rentals are particularly high or low
rentals_by_weekday = day.groupby('weekday')['cnt'].mean().reset_index()

plt.figure(figsize=(10, 4))
plt.bar(rentals_by_weekday['weekday'], rentals_by_weekday['cnt'], color='purple')
plt.xlabel('Day of the Week')
plt.ylabel('Average Number of Rentals')
plt.title('Average Bike Rentals by Day of the Week')
plt.xticks(ticks=range(7), labels=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```python
#8 non-registered contribute to overall bike rentals compared to registered user
# Total rentals
total_rentals = day['cnt'].sum()

# Total rentals by casual users
total_casual_rentals = day['casual'].sum()

# Total rentals by registered users
total_registered_rentals = day['registered'].sum()

# Calculate the percentage contribution
casual_contribution = (total_casual_rentals / total_rentals) * 100
registered_contribution = (total_registered_rentals / total_rentals) * 100

print(casual_contribution)
print(registered_contribution)

# Data for the pie chart
labels = ['Casual Users', 'Registered Users']
sizes = [total_casual_rentals, total_registered_rentals]
colors = ['lightcoral', 'lightskyblue']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=120)
plt.title('Contribution of Casual vs Registered Users to Total Bike Rentals')
plt.show()

```
