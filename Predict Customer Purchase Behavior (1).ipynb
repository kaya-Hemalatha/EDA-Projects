```python
#1 no. of males and females
print("Column names:", my_data.columns)

print("Unique values in 'Gender':", my_data['Gender'].unique())  #used to find the unique values from a series. A series is a single column of a data frame.

num_males = 0
num_female = 0

for gender in my_data['Gender']:
    if (gender== 1):
        num_males += 1
    elif (gender== 0):
        num_female += 1

print(f"number of males in dataset : {num_males}")
print(f"number of females in dataset : {num_female}")

# Bar plot
labels = ['Males', 'Females']
counts = [num_males, num_female]

plt.bar(labels, counts, color=['blue', 'pink'])
plt.xlabel('Count')
plt.ylabel('Gender')
plt.title('Number of Males and Females')
plt.show()
```

```python
#2 number of males and females within the given range
data = my_data[(my_data['Age'] >= 18) & (my_data['Age'] <= 25)]

num_males = (data['Gender'] == 1).sum()
num_females = (data['Gender'] ==0).sum()

print(f"Number of males aged between 18 and 25: {num_males}")
print(f"Number of females aged between 18 and 25: {num_females}")

#pie chart
labels = ['Males', 'Females']
counts = [num_males, num_females]

plt.pie(counts,labels=labels,colors = ['blue','pink'],autopct = '%1.1f%%',startangle = 90)
plt.title('number of males and females within the given range')
plt.show()
```

```python
#3 count of males who purchased the items
#print unique values of productcategory
print("Unique values in Gender:", my_data['ProductCategory'].unique())

categories = [0 , 1 , 2 , 3 , 4]

#filtering the males
males_data = my_data[my_data['Gender'] == 1]

males_categories = males_data[males_data['ProductCategory'].isin(categories)] #used to find the elements contains in a given dataframe,it takes the elements and gets the elements to match the data

category_counts = males_categories['ProductCategory'].value_counts()

for category in categories:
    count = category_counts.get(category) #returns the value associated with the specified key if the key is found in the dictionary.
    print(f"Number of males who purchased {category}: {count}")
    
#plotting graph
category_counts = category_counts.reindex(categories, fill_value=0) #used to change the index of rows and columns of a DataFrame. 
                                                                # used to fill any missing values with the specified value (which is 0 in this case).
#Bar plot
category_counts.plot(kind = 'bar')
plt.xlabel('Product Category')
plt.ylabel('number of males')
plt.title('count of males who purchased each item')
plt.show()
```

```python
#4 Popular Gender who is buying popular products
#popular product
popular_product = my_data['ProductCategory'].value_counts()
print("popular product count", popular_product)

#popular gender in product category
popular_gender = my_data.groupby('Gender')['ProductCategory'].value_counts().unstack()#groupby() allows you to split your data into separate groups to perform computations for better analysis.
print("popular gender count",popular_gender) #unstack() is used to reshape data. It transforms a level of a MultiIndex into columns. 

# Bar plot for overall popularity product category
popular_product.plot(kind='barh', title='Overall Product Category Popularity')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()

#Bar plot for popular gender in product category
popular_gender.plot(kind='bar', title='Popular Gender In Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

```python
#5 Overall most popular product categories
category_counts = my_data['ProductCategory'].value_counts()
print("popular product categories:", category_counts)

# Define age groups
bins = [0, 18, 25, 35, 45, 55, 65, 100]  #Defines the edges of the intervals.
labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+']

my_data['AgeGroup'] = pd.cut(my_data['Age'],bins = bins, labels=labels) #cut() segment and sort data values into bins

# Count the occurrences of each product category by age group
age_category_counts = my_data.groupby('AgeGroup', observed=False)['ProductCategory'].value_counts().unstack()

print("Product category popularity by age group:")
print(age_category_counts)

# Bar plot for age group-specific category popularity
age_category_counts.plot(kind='barh',stacked = 'True', title='Product Category Popularity by Age Group') #stacked Each bar is divided into segments, each representing a different sub-group.
plt.xlabel('Product Category') 
plt.ylabel('Count')
plt.show()
```

```python
#6 customer spends average on product category
spends = my_data.groupby('ProductCategory')['AnnualIncome'].mean().reset_index() #transforms this multi-level index into regular columns, which makes the data easier to work with for visualization and further analysis.
print("average spending in each product category:", spends)

#plotting
sns.barplot(x = 'ProductCategory', y = 'AnnualIncome', data = spends,  hue='ProductCategory', palette = 'viridis') #'viridis' is one of the many built-in color palettes provided by Seaborn.
plt.title('customer spends average value on product category')
plt.xlabel('ProductCategory')
plt.ylabel('Avg AnnualIncome')
plt.show()
```

```python
#7 customer time spent on website
customer_time = my_data['TimeSpentOnWebsite'].describe()
print("overall time spent by customer:", customer_time)

# Box plot by gender
sns.boxplot(x='Gender', y='TimeSpentOnWebsite', data=my_data)
plt.title('Time Spent on Website by Gender')
plt.xlabel('Gender')
plt.ylabel('Time Spent on Website (minutes)')
plt.show()
```

```python
#8 Time spent on website by age
# Create age categories
bins = [18, 25, 35, 45, 55, 65, 100]  #they allow you to group data into intervals or ranges.
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65+']
my_data['AgeGroup'] = pd.cut(my_data['Age'], bins=bins, labels=labels)

# Box plot by age group
sns.boxplot(x='AgeGroup', y='TimeSpentOnWebsite', data=my_data)
plt.title('Time Spent on Website by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Time Spent on Website (minutes)')
plt.show()
```

```python
#9 Does loyalty program making more repeat purchases or not
repeat_purchases = my_data.groupby('LoyaltyProgram')['NumberOfPurchases'].mean()
print("Avg number of purchase:", repeat_purchases)

#visualizing
sns.boxplot(x='LoyaltyProgram', y='NumberOfPurchases', data=my_data)
plt.title('Number of Purchases by Loyalty Program Status')
plt.xlabel('Loyalty Program')
plt.ylabel('Number of Purchases')
plt.show()
```

```python
#10 customer in loyalty program spent more time on website
time_spent = my_data.groupby('LoyaltyProgram')['TimeSpentOnWebsite'].mean()
print("Average time spent on the website:",time_spent)

#plot
sns.boxplot(x='LoyaltyProgram', y='TimeSpentOnWebsite', data=my_data)
plt.title('Time Spent on Website by Loyalty Program Status')
plt.xlabel('Loyalty Program')
plt.ylabel('Time Spent on Website')
plt.show()
```

```python
#11 part of loyalty programs avail more discounts compared to others
average_discounts = my_data.groupby('LoyaltyProgram')['DiscountsAvailed'].mean()
print("Average number of discounts availed by loyalty program members and non-members:", average_discounts)

sns.boxplot(x='LoyaltyProgram', y='DiscountsAvailed', data=my_data)
plt.title('Discounts Availed by Loyalty Program Status')
plt.xlabel('Loyalty Program')
plt.ylabel('Number of Discounts Availed')
plt.show()
```

```python
#12 purchase status by gender
gender_purchase_status = my_data.groupby(['Gender', 'PurchaseStatus']).size().unstack()
print("Purchase Status by Gender:")
print(gender_purchase_status)

# Count plot for purchase status by gender
sns.countplot(x='PurchaseStatus', hue='Gender', data=my_data)
plt.title('Purchase Status by Gender')
plt.xlabel('Purchase Status')
plt.ylabel('Count')
plt.show()
```

```python
#13 purchase status by age
# Create age groups
bins = [0, 25, 35, 45, 55, 65, 100]
labels = ['<25', '25-35', '35-45', '45-55', '55-65', '65+']
my_data['AgeGroup'] = pd.cut(my_data['Age'], bins=bins, labels=labels)

# Group by AgeGroup and PurchaseStatus
age_purchase_status = my_data.groupby(['AgeGroup', 'PurchaseStatus']).size().unstack()
print("Purchase Status by Age Group:", age_purchase_status)

# Count plot for purchase status by age group
sns.countplot(x='PurchaseStatus', hue='AgeGroup', data=my_data)
plt.title('Purchase Status by Age Group')
plt.xlabel('Purchase Status')
plt.ylabel('Count')
plt.show()
```
