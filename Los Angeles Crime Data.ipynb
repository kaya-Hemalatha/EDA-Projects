```python
#1 most common time of day for crimes
# Check the first few rows to understand the format
print(crime['TIME OCC'].head())

# Convert 'TIME OCC' to string if it's not already
crime['TIME OCC'] = crime['TIME OCC'].astype(str)

# Extract the hour from 'TIME OCC' (format "HH:MM")
crime['Hour'] = crime['TIME OCC'].str.split(':').str[0].astype(int)

# Count the occurrences of each hour
time_counts = crime['Hour'].value_counts().sort_index()

# Check if time_counts is empty before proceeding
if not time_counts.empty:
    # Identify the most common time
    most_common_hour = time_counts.idxmax()
    most_common_count = time_counts.max()

    print(f"The most common hour for crimes to occur is: {most_common_hour} with {most_common_count} occurrences.")
else:
    print("No valid time data available for analysis.")
```

```python
#2 highest number of reported crimes
# Group the data by 'AREA NAME' and count the number of crimes in each area
area_counts = crime['AREA NAME'].value_counts()

# Identify the area with the highest number of reported crimes
most_common_area = area_counts.idxmax() #returns the index (area name) of the highest value in area_counts
most_common_area_count = area_counts.max() #returns the highest value itself, i.e., the maximum number of crimes reported in any area

print(f"The area with the highest number of reported crimes is: {most_common_area} with {most_common_area_count} crimes.")

# Plot the data
plt.figure(figsize=(10, 6))  # Set the size of the plot
area_counts.plot(kind='bar', color='skyblue')  # Create a bar plot with light blue bars

plt.title('Number of Reported Crimes by Area')  # Add a title to the plot
plt.xlabel('Area Name')  # Label the x-axis
plt.ylabel('Number of Reported Crimes')  # Label the y-axis
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
#3 most common types of crimes reported
# Group the data by 'Crm Cd Desc' 
crime_counts = crime['Crm Cd Desc'].value_counts()

# Identify the most common crime types
top_crimes = crime_counts.head(10)  # Let's focus on the top 10 most common crime types

print("The most common types of crimes reported are:")
print(top_crimes)

# Pie Chart
plt.figure(figsize=(8, 8))
top_crimes.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(top_crimes)))
plt.title('Top 10 Most Common Types of Crimes (Pie Chart)')
plt.show()
```

```python
#4 part1 and part2 compare the crimes
# Group the data by 'Part 1-2' and count the number of occurrences
part_counts = crime['Part 1-2'].value_counts()

#Print the counts for Part 1 and Part 2 crimes
print("The frequency of Part 1 and Part 2 crimes are:")
print(part_counts)

# 3. Count Plot (Another form of bar plot)
plt.figure(figsize=(8, 6))
sns.countplot(data=crime, x='Part 1-2', palette = 'Set2' , order=part_counts.index)
plt.title('Comparison of Part 1 and Part 2 Crimes (Count Plot)')
plt.xlabel('Crime Category')
plt.ylabel('Number of Occurrences')
plt.show()
```

```python
#5 What is the average age of crime victims

# Calculate the average age of crime victims
average_age = crime['Vict Age'].mean()

# Print the average age
print(f"The average age of crime victims is: {average_age:.2f} years")

# 3. Histogram
plt.figure(figsize=(8, 6))
sns.histplot(crime['Vict Age'], bins=30, kde=True, color='purple')
plt.title('Distribution of Victim Ages (Histogram)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#6 weapons are most frequently used in crimes
# Count the occurrences of each weapon type
weapon_counts = crime['Weapon Desc'].value_counts()

# Print the counts for each weapon type
print("The most frequently used weapons in crimes are:")
print(weapon_counts)

# Horizontal Bar Plot
plt.figure(figsize=(12, 8))
weapon_counts.head(10).plot(kind='barh', color='coral')
plt.title('Top 10 Most Frequently Used Weapons in Crimes (Horizontal Bar Plot)')
plt.xlabel('Number of Occurrences')
plt.ylabel('Weapon Type')
plt.show()

```
